 # Finite Volume Formulation of Linear Convection



 We start from the conservative form:



 $$

 \frac{\partial u}{\partial t} + \frac{\partial f(u)}{\partial x} = 0

 $$



 For linear convection:



 $$

 f(u) = c u

 $$



 ---



 ## Step 1: Integrate Over a Control Volume



 Consider cell $i$:



 $$

 \int_{x_{i-1/2}}^{x_{i+1/2}}

 \left( u_t + f(u)_x \right) dx = 0

 $$



 ---



 ## Step 2: Apply Divergence Theorem



 $$

 \frac{d}{dt} \int u \, dx

 + f(u)_{i+1/2} - f(u)_{i-1/2} = 0

 $$



 ---



 ## Step 3: Define Cell Average



 $$

 \bar{u}_i =

 \frac{1}{\Delta x}

 \int_{x_{i-1/2}}^{x_{i+1/2}} u dx

 $$



 ---



 ## Final Semi-Discrete Finite Volume Form



 $$

 \frac{d \bar{u}_i}{dt}

 =

 - \frac{1}{\Delta x}

 \left(

 f_{i+1/2} - f_{i-1/2}

 \right)

 $$



 The entire problem now reduces to:



 **How do we approximate the numerical flux**

 $f_{i+1/2}$?

 # Central Flux Approximation



 A naive choice:



 $$

 f_{i+1/2}

 =

 \frac{1}{2}

 \left(

 f(u_i) + f(u_{i+1})

 \right)

 $$



 For linear convection:



 $$

 f_{i+1/2}

 =

 \frac{c}{2}

 (u_i + u_{i+1})

 $$



 Substituting into FVM gives:



 $$

 u_i^{n+1}

 =

 u_i^n

 -

 \frac{\text{CFL}}{2}

 (u_{i+1}^n - u_{i-1}^n)

 $$



 This is exactly the **FTCS scheme**.



 ---



  Important:



 This flux contains NO directional information.



 It does not know whether $c > 0$ or $c < 0$.



 Therefore it is unstable.

 ## Energy Analysis



 Define discrete energy:



 $$

 E^n = \sum_i (u_i^n)^2

 $$



 For a stable scheme:



 $$

 E^{n+1} \le E^n

 $$



 FTCS causes energy growth because it lacks dissipation.
